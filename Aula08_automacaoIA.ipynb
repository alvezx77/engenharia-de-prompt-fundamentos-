### Verificação de Temperatura do Sistema (Entrada do Usuário)

O código abaixo agora permite que você insira a temperatura do sistema para demonstrar a lógica de alerta. Ele não simula mais a temperatura.

```python
import time

# Função para obter a temperatura do usuário
def get_user_temperature():
    while True:
        try:
            temp_input = input("Por favor, insira a temperatura atual do sistema em °C (ou 'sair' para encerrar): ")
            if temp_input.lower() == 'sair':
                return None
            temperature = float(temp_input)
            return temperature
        except ValueError:
            print("Entrada inválida. Por favor, digite um número ou 'sair'.")

# Loop para verificar a temperatura continuamente
print("Iniciando monitoramento de temperatura...")
while True:
    current_temperature = get_user_temperature()
    if current_temperature is None:
        break

    print(f"Temperatura inserida: {current_temperature:.2f}°C")

    if current_temperature > 30:
        print("!!! ALERTA: Superaquecimento detectado! Temperatura acima de 30°C !!!")
    else:
        print("Temperatura normal.")

    time.sleep(1) # Pode ajustar o intervalo se desejar

print("Monitoramento de temperatura encerrado.")
```

### Organizador de Arquivos por Extensão

Este programa Python irá varrer um diretório especificado, identificar as extensões dos arquivos (como `.pdf`, `.jpg`, `.txt`, etc.), criar pastas para cada tipo de extensão e, em seguida, mover os arquivos para suas respectivas pastas.

```python
import os
import shutil

def organize_files(directory_path):
    """
    Organiza arquivos em um diretório especificado em subpastas
    baseadas na extensão do arquivo.

    Args:
        directory_path (str): O caminho para o diretório a ser organizado.
    """
    if not os.path.isdir(directory_path):
        print(f"Erro: O diretório '{directory_path}' não existe.")
        return

    print(f"Iniciando organização no diretório: {directory_path}")

    for filename in os.listdir(directory_path):
        file_path = os.path.join(directory_path, filename)

        # Ignorar diretórios e arquivos ocultos ou do sistema
        if os.path.isdir(file_path) or filename.startswith('.'):
            continue

        # Obter a extensão do arquivo
        # os.path.splitext retorna uma tupla (root, ext)
        _, extension = os.path.splitext(filename)

        if extension:
            # Remover o ponto inicial da extensão para usar como nome da pasta
            folder_name = extension[1:].lower()
            destination_folder = os.path.join(directory_path, folder_name)

            # Criar a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print(f"Pasta '{destination_folder}' criada.")

            # Mover o arquivo para a pasta de destino
            try:
                shutil.move(file_path, destination_folder)
                print(f"Movido '{filename}' para '{destination_folder}'")
            except shutil.Error as e:
                print(f"Erro ao mover '{filename}': {e}")
            except Exception as e:
                print(f"Ocorreu um erro inesperado ao mover '{filename}': {e}")
        else:
            print(f"Arquivo '{filename}' sem extensão. Ignorando.")

    print("Organização concluída!")
```

### Consulta de Endereço via ViaCEP

Este programa Python permite consultar informações de endereço utilizando a API pública do ViaCEP. Você irá inserir um CEP e o programa retornará os detalhes do endereço formatados.

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API do ViaCEP para obter informações de endereço.

    Args:
        cep (str): O CEP a ser consultado (apenas dígitos).

    Returns:
        dict or None: Um dicionário com os dados do endereço se a consulta for bem-sucedida,
                      caso contrário, None.
    """
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para códigos de status HTTP ruins (4xx ou 5xx)
        data = response.json()
        if 'erro' in data and data['erro']:
            print(f"Erro: CEP {cep} não encontrado.")
            return None
        return data
    except requests.exceptions.RequestException as e:
        print(f"Erro na requisição: {e}")
        return None

def formatar_endereco(dados_endereco):
    """
    Formata os dados de endereço retornados pela API do ViaCEP de maneira legível.

    Args:
        dados_endereco (dict): Dicionário contendo os dados do endereço.

    Returns:
        str: Uma string formatada com os detalhes do endereço.
    """
    if not dados_endereco:
        return "Nenhum dado de endereço para formatar."

    endereco_formatado = f"CEP: {dados_endereco.get('cep', 'N/A')}\n"
    endereco_formatado += f"Logradouro: {dados_endereco.get('logradouro', 'N/A')}\n"
    endereco_formatado += f"Complemento: {dados_endereco.get('complemento', 'N/A')}\n"
    endereco_formatado += f"Bairro: {dados_endereco.get('bairro', 'N/A')}\n"
    endereco_formatado += f"Localidade: {dados_endereco.get('localidade', 'N/A')}\{dados_endereco.get('uf', 'N/A')}\n"
    endereco_formatado += f"DDD: {dados_endereco.get('ddd', 'N/A')}\n"
    endereco_formatado += f"IBGE: {dados_endereco.get('ibge', 'N/A')}"

    return endereco_formatado
```

#### Exemplo de Uso

Você pode alterar o `cep_para_consultar` para qualquer CEP que desejar. Certifique-se de que o CEP contenha apenas números.

```python
# Exemplo de uso:
cep_para_consultar = input("Digite o CEP (somente números): ") # Exemplo: '01001000'

# Remove qualquer caractere não numérico
cep_para_consultar = ''.join(filter(str.isdigit, cep_para_consultar))

if len(cep_para_consultar) == 8:
    dados = consultar_cep(cep_para_consultar)
    if dados:
        print("\n--- Informações do Endereço ---")
        print(formatar_endereco(dados))
    else:
        print("Não foi possível obter informações para o CEP fornecido.")
else:
    print("Por favor, digite um CEP válido com 8 dígitos numéricos.")
```

### Monitor de Saúde Autônomo com Alertas

Este programa simula um monitor contínuo de saúde do sistema. Ele pedirá que você insira uma pontuação de saúde (de 0 a 100). Se a pontuação cair abaixo de um limite crítico, ele disparará um alerta no console e simulará o envio de um e-mail.

```python
import time

# Limiar crítico de saúde do sistema
CRITICAL_HEALTH_THRESHOLD = 20

def get_system_health_score():
    """
    Solicita ao usuário uma pontuação de saúde do sistema.
    Retorna a pontuação como um inteiro ou None se o usuário quiser sair.
    """
    while True:
        try:
            health_input = input(f"Insira a pontuação de saúde atual (0-100, crítico <= {CRITICAL_HEALTH_THRESHOLD}) ou 'sair' para encerrar: ")
            if health_input.lower() == 'sair':
                return None
            score = int(health_input)
            if 0 <= score <= 100:
                return score
            else:
                print("Pontuação inválida. Por favor, insira um número entre 0 e 100.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro ou 'sair'.")

def send_alert_email(subject, message):
    """
    Simula o envio de um e-mail de alerta.
    """
    print(f"\n--- E-MAIL DE ALERTA SIMULADO ---")
    print(f"Para: equipe_de_monitoramento@empresa.com")
    print(f"Assunto: {subject}")
    print(f"Mensagem: {message}")
    print(f"-----------------------------------\n")

print("Iniciando monitoramento de saúde do sistema...")

while True:
    health_score = get_system_health_score()
    if health_score is None:
        break

    print(f"Pontuação de saúde atual: {health_score}")

    if health_score <= CRITICAL_HEALTH_THRESHOLD:
        alert_message = f"!!! ALERTA CRÍTICO: Saúde do sistema atingiu {health_score}! Limiar crítico é {CRITICAL_HEALTH_THRESHOLD}. !!!"
        print(alert_message)
        send_alert_email("ALERTA CRÍTICO DE SAÚDE DO SISTEMA", alert_message)
    else:
        print("Saúde do sistema normal.")

    time.sleep(2) # Espera 2 segundos antes da próxima verificação (simulado)

print("Monitoramento de saúde do sistema encerrado.")
```
