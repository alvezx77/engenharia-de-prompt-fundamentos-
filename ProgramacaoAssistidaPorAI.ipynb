```python
#SOMA DE LISTAS
numeros = [1,2, 3, 4, 5]
soma = 0
i = 0
while i < len(numeros):
  soma += numeros[i]
  i += 1
print (f"A soma dos números é:{soma} ") #Print não pode estar dentro de "while"
```

```python
#SOMA DE LISTAS REFATORADO
numeros = [1, 2, 3, 4, 5]
soma = sum(numeros)
print(f"A soma dos números é: {soma}")
```

```python
#NÚMERO PRIMO
import math

def eh_primo(n):
  if n < 2:
    return False
  if n == 2:
    return True
  if n % 2 == 0:
    return False
  for i in range(3, int(math.sqrt(n)) + 1, 2):
    if n % i == 0:
      return False
  return True
```

```python
print(f"7 é primo? {eh_primo(7)}")
print(f"10 é primo? {eh_primo(10)}")
print(f"2 é primo? {eh_primo(2)}")
print(f"1 é primo? {eh_primo(1)}") # Note que 1 não é considerado primo pela definição matemática
print(f"13 é primo? {eh_primo(13)}")
print(f"0 é primo? {eh_primo(0)}")
print(f"-5 é primo? {eh_primo(-5)}")
print(f"29 é primo? {eh_primo(29)}")
print(f"35 é primo? {eh_primo(35)}")
```

```python
#INVERSÃO DE STRING
def inverte_texto(texto):
  invertido = ''
  for letra in texto:
    invertido = letra + invertido
  return invertido
```
